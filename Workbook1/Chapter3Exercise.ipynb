```python
x=int (input('Enter number:'))
if x > 50 :
    print('Above 50')
elif x >  20 :
    print('Above 20')
elif x > 10 :
    print('Above 10')
elif x >  0 :
    print('Above 0')
else :
    print('Something else')
```

Exercise 2

```python
hours=float (input('Enter hours worked:'))
rate=float (input('Enter rate:'))

if hours>40:
  pay=40*rate+((hours-40)*(1.5*rate))
  print('Pay:',pay)

else:
  pay=hours*rate
  print('Pay:',pay)
```
